```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('D:\Data.csv')
```

```python
dataset
```

```python
dataset = dataset.dropna(axis=1)
```

```python
#Count of Malignant(M) and Benign(B)
dataset['diagnosis'].value_counts()
```

```python
#Encoding the categorical data 
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
dataset.iloc[:,1] = labelencoder_Y.fit_transform(dataset.iloc[:,1].values)
print(labelencoder_Y.fit_transform(dataset.iloc[:,1].values))
```

```python
#Splitting data into independent(X) and dependent(Y) datasets
X = dataset.iloc[:,2:31].values
Y = dataset.iloc[:,1].values
```

```python
#Splitting data into train:test ratio 70:30
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.30,random_state=0)
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
X_train = ssc.fit_transform(X_train)
X_test = ssc.fit_transform(X_test)
```

```python
def models(X_train,Y_train):
  
  #Using KNeighborsClassifier(KNN) 
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    knn.fit(X_train, Y_train)

  #Using Naive Bayes 
    from sklearn.naive_bayes import GaussianNB
    gauss = GaussianNB()
    gauss.fit(X_train, Y_train)

  #Using DecisionTreeClassifier 
    from sklearn.tree import DecisionTreeClassifier
    destree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    destree.fit(X_train, Y_train)

  #Using Random Forest Classifier 
    from sklearn.ensemble import RandomForestClassifier
    fore = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    fore.fit(X_train, Y_train)
  
    print('0.K Nearest Neighbor Training Accuracy:', knn.score(X_train, Y_train))
    print('1.Gaussian Naive Bayes Training Accuracy:', gauss.score(X_train, Y_train))
    print('2.Decision Tree Classifier Training Accuracy:', destree.score(X_train, Y_train))
    print('3.Random Forest Classifier Training Accuracy:', fore.score(X_train, Y_train))
  

    return  knn, gauss, destree, fore
```

```python
model = models(X_train,Y_train)
```

```python
#test model accuracy 


from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    
    print('Model ',i)
    com = confusion_matrix(Y_test, model[i].predict(X_test))
  
    TP = cm[0][0]
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]
  
    print(com)
    print('Testing Accuracy = ', (TP + TN) / (TP + TN + FN + FP))
    print()
```

```python

#We have now created a breast detection program to predict if a patient has cancer or not.According to the data above, the most 
#accurate value was provided by Random Forest Classifier algorithm with accuracy of 95.90%. Hence, the best suited algorithm for 
#this task is the Random Forest Classifier algorithm.
```
